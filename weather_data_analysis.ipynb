```python
import pandas as pd 
```

```python
df_weather = pd.read_csv('weather_data.csv')
df_weather
```

```python
## listing out all the names of columns

column = df_weather.columns
column

```

```python
## Find all the unique 'Wind Speed' values in the data
## set() function is used to extract unique values from column 'Wind Speed_km/h'

unique_values_set = set(df_weather['Wind Speed_km/h'])
unique_values_set

```

```python
## number of times when the Weather is exactly Clear

weathers = df_weather.get('Weather')
weathers.value_counts()['Clear']
```

```python
##  number of times when the 'Wind Speed was exactly 4 km/h

wind= df_weather.get('Wind Speed_km/h')
wind.value_counts()[4]
```

```python
## Find out all the Null Values in the data

df_weather.isnull().sum()
```

```python
# Rename the column name 'Weather' of the dataframe to 'Weather Condition'

df_weather.rename(columns = {"Weather" : "Weather Condition"})
```

```python
## Find mean 'Visibility'

means = df_weather.get('Visibility_km')
means.mean().round(2)
```

```python
## Standard Deviation of Pressure

SD_press = df_weather.get('Press_kPa')
SD_press.std().round(2)
```

```python
## finding variance of Relative humidity

rel_hum= df_weather.get('Rel Hum_%')
rel_hum.var().round(2)
```

```python
## all instances when 'Snow' was recorded
## equality operator ( == ) 

df_snow = df_weather[df_weather['Weather'] == 'Snow']
df_snow
```

```python
#extracting the information using loc (weather = snow)

df_snow = df_weather.loc[(df_weather.Weather == 'Snow')]
df_snow
```

```python
## all instances when 'Wind Speed is above 24' and 'Visibility is 25'

df_wind = df_weather[(df_weather ['Wind Speed_km/h'] > 24) & (df_weather['Visibility_km'] == 25)]
df_wind

```

```python
##  Find all instances when 'Weather is Clear' or 'Visibility is above 40'

weather_visibility= df_weather[(df_weather['Weather'] == 'Mainly Clear') | (df_weather['Visibility_km'] > 40)]
weather_visibility
```

```python
## Find all instances when :
# A. 'Weather is Clear' and 'Relative Humidity is greater than 50'
# or
# B. 'Visibility is above 40'

last = df_weather[((df_weather['Weather'] == 'Mainly Clear') & (df_weather['Rel Hum_%']>50) | (df_weather['Visibility_km'] > 40))]
last
```

```python
##  the Minimum value of each column against each Weather

df_dropped = df_weather.drop('Date/Time', axis='columns')
df_dropped.groupby(by= 'Weather').min()

```

```python
##  the Maximum value of each column against each Weather

df_dropped.groupby(by='Weather').max()
```

```python
## the Mean value of each column against each 'Weather'

df_dropped.groupby(by = 'Weather').mean().round(2)
```
